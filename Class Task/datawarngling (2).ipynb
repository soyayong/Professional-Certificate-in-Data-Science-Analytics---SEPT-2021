### Data Formatting

## Data Normalization

```python
df[['width', 'height', 'length']].head(20)
```

```python
df['width'] = df['width']/df['width'].max()

df['height'] = df['height']/df['height'].max()

df['length'] = df['length']/df['length'].max()
```

## Get Dummies

```python
dummies_fueltype = pd.get_dummies(df['fuel-type'])

dummies_fueltype.tail(30)
```

```python
dummies_fueltype.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)

dummies_fueltype.head()
```

```python
dfnew = pd.concat([df, dummies_fueltype], axis=1)

dfnew.head()
```

```python
dfnew.drop('fuel-type', axis=1, inplace=True)

dfnew.head()
```

```python
dfnew.describe(include='all')
```

```python
dfnew['aspiration'].value_counts()
```

```python
dummies_aspiration = pd.get_dummies(df['aspiration'])

dummies_aspiration.tail(30)
```

```python
dummies_aspiration.rename(columns={'std':'aspiration-std', 'turbo':'aspiration-turbo'}, inplace=True)

dummies_aspiration.head()
```

```python
df2 = pd.concat([dfnew, dummies_aspiration], axis=1)

df2.head()
```

## Binning

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df2["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
categories_names = ['Low' , 'Medium' , 'High']
```

```python
df2['horsepower-binned'] = pd.cut(df2['horsepower'], bins, labels=categories_names, include_lowest=True )
df2[['horsepower','horsepower-binned']].head(30)
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df2["price"])

# set x/y labels and plot title
plt.pyplot.xlabel("price")
plt.pyplot.ylabel("count")
plt.pyplot.title("price bins")
```

```python
bins = np.linspace(min(df2["price"]), max(df2["price"]), 4)
bins
```

```python
df2['price-binned'] = pd.cut(df2['price'], bins, labels=categories_names, include_lowest=True )
df2[['price','price-binned']].head(30)
```
