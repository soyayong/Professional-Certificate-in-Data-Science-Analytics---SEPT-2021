```python
# Group by
df2 = df[['drive-wheels', 'body-style', 'price']]

df2_group = df2.groupby(['drive-wheels', 'body-style'], as_index= False).mean()

df2_group
```

```python
pivot2 = df2_group.pivot(index='drive-wheels', columns='body-style')

pivot2
```

```python
df2 = df[['drive-wheels', 'body-style', 'price']]

df2_group = df2.groupby(['drive-wheels', 'body-style'], as_index= True).mean()

df2_group
```

```python
df3 = df[['drive-wheels', 'num-of-doors', 'price']]

df3_group = df3.groupby(['drive-wheels', 'num-of-doors'], as_index= False).mean()

df3_group
```

```python
pivot3 = df3_group.pivot(index='drive-wheels', columns='num-of-doors')

pivot3
```

```python
#Heatmap
import numpy as np


fig, ax = plt.subplots()
im = ax.pcolor(pivot2, cmap='RdBu')

#label names
row_labels = pivot2.columns.levels[1]
col_labels = pivot2.index

#move ticks and labels to the center
ax.set_xticks(np.arange(pivot2.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(pivot2.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

```python
fig, ax = plt.subplots()
im = ax.pcolor(pivot3, cmap='YlGn')

#label names
row_labels = pivot3.columns.levels[1]
col_labels = pivot3.index

#move ticks and labels to the center
ax.set_xticks(np.arange(pivot3.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(pivot3.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

```python
# Person Method

from scipy import stats

pearson_coef , p_value = stats.pearsonr(df["engine-size"], df["price"])
print("Pearson Correlation is", pearson_coef, 'P_Value_is', p_value)
```
