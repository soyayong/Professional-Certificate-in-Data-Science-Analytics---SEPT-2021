## Binning 

```python
df4.head()
```

```python
df4['horsepower'].head(30)
```

```python
df4['normalized-losses'].head(30)
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df4["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
bins = np.linspace(min(df4["horsepower"]), max(df4["horsepower"]), 4)
bins
```

```python
categories_names = ['Low' , 'Medium' , 'High']
```

```python
df4['horsepower-binned'] = pd.cut(df4['horsepower'], bins, labels=categories_names, include_lowest=True )
df4[['horsepower','horsepower-binned']].head(30)
```

```python
df4[['horsepower','horsepower-binned']].tail(30)
```

```python
df4['horsepower-binned'].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df4["price"])

# set x/y labels and plot title
plt.pyplot.xlabel("price")
plt.pyplot.ylabel("count")
plt.pyplot.title("price bins")
```

```python
bins = np.linspace(min(df4["price"]), max(df4["price"]), 4)
bins
```

```python
df4['price-binned'] = pd.cut(df4['price'], bins, labels=categories_names, include_lowest=True )
df4[['price','price-binned']].head(30)
```

```python
df4['price-binned'].value_counts()
```
